```python
linnarson_DUB_EXC = pd.read_csv('data_sensory/Region_EXC_PNS.csv',index_col=0)
barbara_EXC = pd.read_csv('data/barbara_EXC.csv',index_col=0)

base = linnarson_DUB_EXC.drop('Region',axis = 1)

labels = linnarson_DUB_EXC['Region'].values

base = base[base.columns & barbara_EXC.columns]
base = base.loc[:,barbara_EXC.columns.values.tolist()]

scaler = StandardScaler()

base_scaled = scaler.fit_transform(base)
```

```python
base
```

```python
# Classify the new dataset using the best models


all_data_DUB_scaled = scaler.fit_transform(barbara_EXC)

classifiers = [
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    GaussianNB(),
    SVC(),
    MLPClassifier()
]

models = [

    KNeighborsClassifier(n_neighbors=15),
    DecisionTreeClassifier(max_depth=10,min_samples_split=10),
    GaussianNB(priors=None, var_smoothing=1e-9),
    SVC(kernel='linear'),
    MLPClassifier(activation='tanh', 
                   hidden_layer_sizes = (837, 8), 
                   learning_rate_init = 0.01, 
                   max_iter = 450, 
                   solver = 'sgd')
]

prediction = []
for classifier, model in zip(classifiers, models):
    model_predictions = model.fit(base_scaled,labels)
    y_pred = model_predictions.predict(all_data_DUB_scaled)
    prediction.append(y_pred)
    
    print("Classifier:", classifier.__class__.__name__)
    print("------------------------ Done")
```

```python
barbara_EXC['kNN'] = prediction[0].tolist()
barbara_EXC['DT'] = prediction[1].tolist()
barbara_EXC['NB'] = prediction[2].tolist()
barbara_EXC['SVM'] = prediction[3].tolist()
barbara_EXC['MLP'] = prediction[4].tolist()
```

```python
prediction = ['kNN','DT','NB','SVM','MLP']

for i in range(0,len(prediction)):
    print(barbara_EXC.groupby(prediction[i]).size())
    print('\n')
    
```
